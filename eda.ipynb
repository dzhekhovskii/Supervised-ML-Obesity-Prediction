# **Numerical features analysis**

```python
# Select only numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Numerical features:", numeric_cols)

# 1. Show descriptive statistics
print("\n Descriptive Statistics for Numerical Features")
display(df[numeric_cols].describe())


# 2. Histograms for each numerical feature
print("\n Histograms for Numerical Features")
for col in numeric_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.grid(False)
    plt.show()


# 3. Boxplots for each numerical feature
print("\n Boxplots for Numerical Features")
for col in numeric_cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.grid(False)
    plt.show()
```

# **Categorical features analysis**

```python
# Select only categorical columns (excluding target column)

TARGET_COL = "NObeyesdad"

categorical_cols = [
    col for col in df.select_dtypes(include=['object']).columns
    if col != TARGET_COL
]

print("Categorical features (excluding target column):", categorical_cols)


# Show descriptive statistics for categorical features
print("\n Descriptive Statistics for Categorical Features")
for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].describe())
    print("Unique categories:", df[col].unique())

# Countplots for each categorical feature
print("\n Countplots for Categorical Features")
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=df[col])
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(False)
    plt.show()


# 3. Proportions of each category
print("\n Category Proportions")
for col in categorical_cols:
    print(f"\n Proportions for {col}")
    print(df[col].value_counts(normalize=True).round(3))
```

# **Target Variable Analysis**

```python
print(f"Target variable: {TARGET_COL}")

# 1. Show basic statistics
print("\n Target Variable Overview")
print(df[TARGET_COL].describe())
print("Unique classes:", df[TARGET_COL].unique())


# 2. Countplot of class distribution
plt.figure(figsize=(8,4))
sns.countplot(x=df[TARGET_COL])
plt.title("Class Distribution of Target Variable")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(False)
plt.show()


# 3. Show proportion of each class
print("\n Class Proportions")
class_proportions = df[TARGET_COL].value_counts(normalize=True).round(3)
print(class_proportions)
```

# **Correlation matrix and Variance**

```python
# CORRELATION MATRIX
print("\n Correlation Matrix")
corr_matrix = df[numeric_cols].corr()
display(corr_matrix)

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

# **Cleaning**

```python
# Drop duplicates
df = df.drop_duplicates()

# Drop irrelevant columns
df = df.drop(columns=['SMOKE', 'SCC', "FCVC", "TUE", 'Weight', 'Height'])

# Water CH2O: round to 2 decimals
df["CH2O"] = df["CH2O"].round(2)

# Other numeric columns to integer
int_columns = ["Age", "NCP", "FAF"]
df[int_columns] = df[int_columns].astype(int)
```

# Exploratory Data Analysis (EDA)  Obesity Dataset

## 1. Dataset Overview
This dataset contains information on individuals from Mexico, Peru, and Colombia to estimate obesity levels based on eating habits, lifestyle, and physical condition.
- Total records: 2,111
- Total features: 17 (demographic, behavioral, physiological)
- Target variable: NObeyesdad (Obesity Level)
- Note: 77% of data was synthetically generated using SMOTE, 23% collected from real users.

## 2. Numerical Features Analysis
- Numerical features: `['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']`
- Calculated descriptive statistics: mean, std, min, max, quartiles
- Variance analysis: identified low-variance features
- Correlation matrix + heatmap
- Observations:
    - Height and Weight strongly correlated (~0.46)
    - Weight moderately correlated with Age (~0.18)
    - Weak negative correlation between Age and FAF (~ -0.16)
    - Most other correlations are weak, favorable for ML

## 3. Categorical Features Analysis
- Categorical features: `['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']`
- Countplots and category proportions
- Observations:
    - SMOKE (2.1% yes) and SCC (4.5% yes) have very low variance
    - Categories unbalanced in CAEC, CALC, MTRANS
    - Insights will guide encoding in ML pipeline

## 4. Target Variable Analysis
- NObeyesdad: 7 original classes
- Class proportions roughly balanced
- Plan for ML: merge into 4 simplified classes: `Insufficient weight`, `Normal weight`, `Overweight`, `Obesity`

## 5. Data Cleaning for Analysis
- Removed 24 duplicate records
- Dropped columns irrelevant for analysis: `SMOKE`, `SCC`, `FCVC`, `TUE`, `Weight`, `Height`

    `SMOKE`  very low variance (only ~2% of individuals smoke), unlikely to contribute meaningful signal for obesity prediction.

    `SCC`  very low variance (only ~4.5% monitor calorie intake), not informative for ML modeling.

    `FCVC`  unclear meaning; the column asks Do you usually eat vegetables in your meals? but values are inconsistent and ambiguous.

    `TUE`  daily technology use showed no clear relationship with obesity.

    `Weight` & `Height`  these features are highly correlated with obesity.

- Converted numeric columns to appropriate types for analysis (`Age`, `NCP`, `CH2O`, `FAF`)


## 6. Summary & Next Steps
- Dataset cleaned for analysis: duplicates removed, numeric types converted, irrelevant columns dropped
- Identified which features are categorical/numerical and their distributions
- Observations will guide preprocessing and feature engineering in the ML pipeline
